# Statistical Thinking Notebook

This notebook will be an example of data exploratory analysis in Python.

```python
# Import plotting modules
import matplotlib.pyplot as plt
import seaborn as  sns
import pandas as pd
import os
import numpy as np
# Set default Seaborn style
sns.set()
fifadatafile = os.path.join(os.getcwd(),'data','fifa19','data.csv')
fifadata = pd.read_csv(fifadatafile)
```

```python
spanishteamsfile = os.path.join(os.getcwd(),'data','fifa19','teams.csv')
spanishteams = pd.read_csv(spanishteamsfile)

spafifadata= pd.merge(fifadata, spanishteams, how='inner', left_on='Club', right_on='Club',
         left_index=False, right_index=False, sort=True,copy=True, indicator=False,
         validate=None)
```

Following code will present age histograma analysis of all players by age of Spanish League.

```python
# Printing Histogram with Seaborn imported library: Using Age column -->
# Compute number of data points: n_data
n_bins=len(spafifadata.Age.value_counts())
# Convert number of bins to integer: n_bins
n_bins=int(n_bins)
_=plt.hist(spafifadata.Age,bins=n_bins)
# Label axes
_=plt.xlabel('Age (Years)')
_=plt.ylabel('count')
# Show histogram
plt.show()
```

```python
# Create bee swarm plot with Seaborn's default settings
fig= plt.figure(figsize=(36,16))
_=sns.swarmplot(x='Club',y='Age',data=spafifadata)

# Label the axes
_ = plt.xlabel('Club')
_ = plt.ylabel('Age')


# Show the plot
plt.show()
```
